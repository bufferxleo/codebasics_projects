we need a function that takes photo as input and returns the cropped roi

```python
original_image=cv2.imread('./dataset/lionel_messi/02453493f852c0efe8ff951dea5845c4.jpg')
plt.imshow(original_image)
```

```python
def get_cropped_image_if_2eyes(img_path):
    img=cv2.imread(img_path)
    if(img is not None):
        gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
        faces=face_cascade.detectMultiScale(gray,1.3,5)
        for (x,y,w,h) in faces:
            roi_gray=gray[y:y+h,x:x+h]
            roi_color=img[y:y+h,x:x+w]
            eyes=eye_cascade.detectMultiScale(roi_gray)
            if len(eyes)==2:
                return roi_color
```

```python
original_image=cv2.imread('./dataset/lionel_messi/02453493f852c0efe8ff951dea5845c4.jpg')
plt.imshow(original_image)
```

```python
cropped_img=get_cropped_image_if_2eyes('./dataset/lionel_messi/02453493f852c0efe8ff951dea5845c4.jpg')
cropped_img
plt.imshow(cropped_img)
```

if 2 eyes are not clearly traced then the fuction wil return nothing

```python
path_to_data='./dataset/'
path_for_cropped_data='./dataset/cropped/'
```

```python
import os
img_dirs=[]
for entry in os.scandir(path_to_data):
    if entry.is_dir():    
        img_dirs.append(entry.path)
img_dirs
```

create cropped folder

```python
import shutil
if os.path.exists(path_for_cropped_data):
    shutil.rmtree(path_for_cropped_data)
os.mkdir(path_for_cropped_data)
```

```python
cropped_image_dirs = []
celebrity_file_names_dict = {}
for img_dir in img_dirs:
    count = 1
    celebrity_name = img_dir.split('/')[-1]
    celebrity_file_names_dict[celebrity_name] = []
    for entry in os.scandir(img_dir):
        roi_color = get_cropped_image_if_2eyes(entry.path)
        if roi_color is not None:
            cropped_folder = path_for_cropped_data + celebrity_name
            if not os.path.exists(cropped_folder):
                os.makedirs(cropped_folder)
                cropped_image_dirs.append(cropped_folder)
                print("Generating cropped images in folder: ",cropped_folder)
            cropped_file_name = celebrity_name + str(count) + ".png"
            cropped_file_path = cropped_folder + "/" + cropped_file_name
            cv2.imwrite(cropped_file_path, roi_color)
            celebrity_file_names_dict[celebrity_name].append(cropped_file_path)
            count += 1
```
