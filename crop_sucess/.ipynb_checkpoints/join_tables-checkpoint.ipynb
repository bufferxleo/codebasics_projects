```python
l=[]
for i in range(len(df2)):
    try:
        if df2['label'][i]=='rice':
            l.append(0)
        elif df2['label'][i]=='cotton':
            l.append(1)
        else:
            l.append(np.nan)
            
    except:
        pass
```

```python
df2['crop']=l
df2['crop'].unique()
```

```python
df2=df2.dropna()

df2.shape
```
